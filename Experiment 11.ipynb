```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Sample Text Data
texts = [
    "Natural Language Processing is very interesting and useful.",
    "NLTK is a powerful library for text processing.",
    "SpaCy is fast and efficient for NLP tasks.",
    "Machine learning models require clean and processed data.",
    "Tokenization, stemming, and lemmatization are important steps in NLP."
]

# Create Stemmer
ps = PorterStemmer()

# Tokenization and Stemming
for sentence in texts:

    # Tokenization
    tokens = word_tokenize(sentence)

    # Stemming
    stemmed_words = [ps.stem(word) for word in tokens]

    print("Original Sentence:")
    print(sentence)

    print("\nTokens:")
    print(tokens)

    print("\nStemmed Words:")
    print(stemmed_words)

    print("\n" + "="*60 + "\n")
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download required resources
nltk.download('stopwords')
nltk.download('wordnet')

# Create Objects
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Processing Text
for sentence in texts:

    # Tokenization
    tokens = word_tokenize(sentence)

    # Remove Stopwords
    filtered_words = [
        word for word in tokens
        if word.lower() not in stop_words
    ]

    # Lemmatization
    lemmatized_words = [
        lemmatizer.lemmatize(word)
        for word in filtered_words
    ]

    print("Original Sentence:")
    print(sentence)

    print("\nAfter Stopword Removal:")
    print(filtered_words)

    print("\nLemmatized Words:")
    print(lemmatized_words)

    print("\n" + "="*60 + "\n")
```

```python
import spacy

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

print("SpaCy Loaded Successfully")

# Sample Text Data
texts = [
    "Natural Language Processing is very interesting and useful.",
    "NLTK is a powerful library for text processing.",
    "SpaCy is fast and efficient for NLP tasks.",
    "Machine learning models require clean and processed data.",
    "Tokenization, stemming, and lemmatization are important steps in NLP."
]

# Process Text Data
for sentence in texts:

    # Create document
    doc = nlp(sentence)

    # Remove stopwords and punctuation + Lemmatization
    processed_words = [
        token.lemma_
        for token in doc
        if not token.is_stop and not token.is_punct
    ]

    print("Original Sentence:")
    print(sentence)

    print("\nProcessed Words (SpaCy):")
    print(processed_words)

    print("\n" + "="*60 + "\n")
```
