```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

model = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(units=1)
])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.1), loss='mean_squared_error')
model.fit(celsius, fahrenheit, epochs=1000, verbose=0)

test_value = np.array([[25.0]])
prediction = model.predict(test_value)

print(f"25°C = {prediction[0][0]:.2f}°F")
```
